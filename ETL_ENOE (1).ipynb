La siguiente sección se comienza la lectura de los archivos y el procesado de información con su respectivo dataframe.
Esta sección incluye lectura, ciclo automatizado y comparación de entidades extientes con la base de datos actual

```python
# Importamos las librerías de Python necesarias para realizar la limpieza
import os
import re
import pandas as pd
import numpy as np

Tablas_limpias = []
# Agregamos los archivos a limpiar, carpeta a limpiar, en este caso leeremos aproximadamente 40 archivos por carpeta
# Guardaremos las rutas de lecturas de los archivos, el archivo que contiene los registros de las entidades y los copararemos con cada uno existente
carpeta = '121_enoe_n_indicadores_estrategicos_2022_trim1_xls/Ciudades'

# Ruta donde se guardarán los archivos CSV
ruta_guardado = carpeta

# Tabla de las entidades
Tb_ciudades = pd.read_excel('Ciudades_sin_duplicados.xlsx')

```

Una vez leidos, comparados con la DB existente y definido la estructura del dataframe, procedemos a cargar cada dataframe para analizar que todos fuesen procesados correctamente, llamando cada tabla por la entidad que correspone

```python
# Lista para almacenar los nombres de los municipios
municipios = []

# Iterar sobre los archivos en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith('.xls') or archivo.endswith('.xlsx'):
        # Obtener el nombre del municipio del archivo
        nombre_sin_extension = archivo.split('.')[0]
        municipio = nombre_sin_extension.split('_')[4]  # Obtener el nombre del municipio
        municipios.append(municipio)  # Agregar el nombre del municipio a la lista

# Diccionario para almacenar las tablas limpias por municipio
dict_tablas_por_municipio = {}

# Iterar sobre las tablas limpias y asociarlas con su municipio correspondiente en el diccionario
for municipio, tabla in zip(municipios, Tablas_limpias):
    dict_tablas_por_municipio[municipio] = tabla

# Imprimir cada tabla resultante
for municipio, tabla in dict_tablas_por_municipio.items():
    print(f"Tabla para el municipio '{municipio}':")
    print(tabla)
    print("\n")
```

Una vez caragados y tratados los dataframes, procedemos a la exportación en csv, este proceso será rapido, ya que anteriormente definimos las rutas y los nombres de los archivos a expotar

```python
import os
# Codigo para exportar los datos a csv

# Iterar sobre las tablas limpias y exportarlas a archivos CSV
for municipio, tabla in zip(municipios, Tablas_limpias):
    # Generar el nombre del archivo CSV
    nombre_archivo = f"Enoe_4_Entidad_{municipio}_Trabajadores-subordinados-y-remunerado-por_Miguel-Angel-Ubaldo-Villalobos.csv"

    # Ruta completa del archivo CSV
    ruta_completa = os.path.join(ruta_guardado, nombre_archivo)

    # Exportar la tabla a un archivo CSV
    tabla.to_csv(ruta_completa, index=False)

    # Imprimir mensaje de confirmación
    print(f"Tabla para el municipio '{municipio}' exportada como '{ruta_completa}'")
```

Para este exemplo se utilizó la primera carpeta a extraer, dando como resultado  un aproximado de 36,400 registros procesados en esta unica carpeta, añadiendo las demás carpetas dan como un resultado de 200 mil regusistros aproximados
